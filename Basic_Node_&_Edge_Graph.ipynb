**1. Buat node dan edge**

```
import networkx as nx
```

```
Graf = nx.Graph()
```

```
Graf.add_nodes_from([1,2,3,4,5,6,7,8])
Graf.add_edges_from([
                  (1,2),
                  (1,3),
                  (1,4),
                  (1,5),
                  (1,6),
                  (1,7),
                  (1,8),
                  (5,6),
                  (5,8),
                  (6,7),])
```

```
nx.draw_shell(Graf, with_labels=True)
```

2.a. Menghapus node 4

```
Graf.remove_node(4)
nx.draw_shell(Graf, with_labels=True)
```

2.b. Menghapus edge (1,8)

```
Graf.remove_edge(1,8)
nx.draw_shell(Graf, with_labels=True)
```
